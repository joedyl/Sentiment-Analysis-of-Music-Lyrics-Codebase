Some additional checks due to initial errors when running data through agreement file

```python
# additional checks due to errors calculating alpha

sinan_x = sinan['x_coordinate'].tolist()
sinan_y = sinan['y_coordinate'].tolist()

sam_x = samuel['x_coordinate'].tolist()
sam_y = samuel['y_coordinate'].tolist()

for i, coord in enumerate(sam_x):
    if coord == '--0.8':
        print(f"Error caused by index {i}")

sam_x[186] = '-0.8'

for i, coord in enumerate(sam_y):
    if coord == '-.0.5':
        print(f"Error caused by index {i}")


sam_y[46] = '-0.5'

samuel['x_coordinate'] = sam_x

samuel['y_coordinate'] = sam_y
 

# rounding up values to make an even number of coords, granularity was meant to be 0.1

for i, value in enumerate(sinan_x):
    if value == 0.25:
        sinan_x[i] = 0.3
    if value == 0.85:
        sinan_x[i] = 0.9
    if value == -0.55:
        sinan_x[i] = -0.6
    if value == 0.75:
        sinan_x[i] = 0.8
    if value == -0.75:
        sinan_x[i] = -0.8
    if value == 0.65:
        sinan_x[i] = 0.7
    if value == 0.19:
        sinan_x[i] = 0.2

for i, value in enumerate(sinan_y):
    if value == 0.65:
        sinan_y[i] = 0.7
    if value == 0.85:
        sinan_y[i] = 0.9
    if value == 0.55:
        sinan_y[i] = 0.6
    if value == 0.45:
        sinan_y[i] = 0.5
    if value == -0.35:
        sinan_y[i] = -0.4
    if value == -0.15:
        sinan_y[i] = -0.2


sinan['x_coordinate'] = sinan_x

sinan['y_coordinate'] = sinan_y


```

Converting to float as they are strings, for ease of agreement calculation

```python
sam_x = samuel['x_coordinate'].tolist()
sam_y = samuel['y_coordinate'].tolist()

for i, value in enumerate(sam_x):
    sam_x[i] = float(value)

for i, value in enumerate(sam_y):
    sam_y[i] = float(value)

samuel['x_coordinate'] = sam_x

samuel['y_coordinate'] = sam_y
```

Making plots from recorded alpha/beta statistics

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

x_beta = [0.65, 0.70, 0.67, 0.68, 0.71, 0.68]

y_beta = [0.63, 0.63, 0.57, 0.60, 0.61, 0.62]

beta = pd.DataFrame({'x_beta': x_beta, 'y_beta': y_beta}, 
                  index=['rap', 'rnb', 'pop', 'rock', 'country', 'all_genres'])

x_alpha = [0.33, 0.40, 0.35, 0.36, 0.42, 0.38]

y_alpha = [0.26, 0.27, 0.15, 0.22, 0.22, 0.25]

alpha = pd.DataFrame({'x_alpha': x_alpha, 'y_alpha': y_alpha},
                     index=['rap', 'rnb', 'pop', 'rock', 'country', 'all_genres'])
```

```python
# Initialize the figure
plt.figure(figsize= (8, 8))

# Plot each dataset with different markers and colors
plt.scatter(alpha['x_alpha'], alpha['y_alpha'], label='Alpha', marker='o', color='blue')
plt.scatter(beta['x_beta'], beta['y_beta'], label='Beta', marker='s', color='green')


# Adding labels for each point
for label, x, y in zip(alpha.index, alpha['x_alpha'], alpha['y_alpha']):
    plt.text(x, y, label, fontsize=9)

for label, x, y in zip(beta.index, beta['x_beta'], beta['y_beta']):
    plt.text(x, y, label, fontsize=9)

# Set title and labels
plt.title('Comparative Scatter Plot of Different Metrics')
plt.xlabel('X')
plt.ylabel('Y')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
