Performing checks to see where agreement might land on first 100 annotations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

samudp ='samucomp.csv'
sinandp = 'sinancomp.csv'

samuds = pd.read_csv(samudp)
sinands = pd.read_csv(sinandp)
joeds = pd.read_csv('joecomp.csv')
```

```python
sinan_x = sinands[["index", "x_coordinate", "y_coordinate"]].head(100)



print(sinan_x)

samu_x = samuds[["index", "x_coordinate", "y_coordinate"]].head(100)



print(samu_x)

joe_x = joeds[["Unnamed: 0", "x_coordinate", "y_coordinate"]].head(100)
print(joe_x)



```

```python
print("Sinan: \n")
print(sinan_x.describe())
print("Sam: \n")
print(samu_x.describe())
print("Joe: \n")
print(joe_x.describe())
```

```python
def check_points_across_dataframes(df1, df2, df3, range_limit):
    count = 0
    num_points = len(df1)
    
    for i in range(num_points):
        # Get coordinates for the current point from all three dataframes
        x1, y1 = df1.iloc[i]['x_coordinate'], df1.iloc[i]['y_coordinate']
        x2, y2 = df2.iloc[i]['x_coordinate'], df2.iloc[i]['y_coordinate']
        x3, y3 = df3.iloc[i]['x_coordinate'], df3.iloc[i]['y_coordinate']
        
        # Calculate pairwise distances
        dist_12 = np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
        dist_13 = np.sqrt((x1 - x3) ** 2 + (y1 - y3) ** 2)
        dist_23 = np.sqrt((x2 - x3) ** 2 + (y2 - y3) ** 2)
        
        # Check if all distances are within the range limit
        if dist_12 <= range_limit and dist_13 <= range_limit and dist_23 <= range_limit:
            count += 1
    
    return count
```

```python
range_limit = 0.4
count_within_range = check_points_across_dataframes(samu_x, joe_x, sinan_x, range_limit)
print(f"Number of points within {range_limit} across the dataframes: {count_within_range}")

```
