Performing basic checks on original data sample - sampled_genius_lyrics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_path = 'sampled_genius_lyrics.csv'
data = pd.read_csv(data_path)

data
```

```python
data['language'].value_counts()
```

The language is all the same, so this column is dropped

```python
data['tag'].value_counts()
```

```python
misc = data[data['tag'] == 'misc']

misc
```

```python
data = data.drop(columns=['views', 'features', 'id', 'language_cld3', 'language_ft', 'language'])
```

```python
data = data.drop(index=range(1415, 1468), axis=0)
```

Misc is not music, its other things like books, so those indexes are found through the misc subset and dropped
Also columns not of any use to us such as id is dropped. 

```python
data.reset_index(inplace=True)
```
